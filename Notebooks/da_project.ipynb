```python
#visual comparision of date and congestion level
import matplotlib.pyplot as plt
import pandas as pd


a = df.groupby(df['Date'].dt.to_period('M'))[['Congestion Level']].mean().reset_index()


a['Date'] = a['Date'].dt.to_timestamp()


plt.figure(figsize=(10,6))
plt.plot(a['Date'], a['Congestion Level'], marker='o', color='blue')
print(a)

plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Average Congestion Level (%)")
plt.title("Monthly Average Congestion Level Over Time")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


corr = df.corr(numeric_only=True)  
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numeric Columns", fontsize=14)
plt.show()
```

```python
#finding distribution of congestion
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
# Plot histogram without KDE first
sns.histplot(df['Congestion Level'], bins=20, color='skyblue', edgecolor='black', stat='density')

# Add KDE curve separately with custom color
sns.kdeplot(df['Congestion Level'], color='red', linewidth=2)

plt.title('Distribution of Congestion Level with KDE Curve')
plt.xlabel('Congestion Level (%)')
plt.ylabel('Density')
plt.show()
```
